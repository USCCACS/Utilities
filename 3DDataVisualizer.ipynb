```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np

nx=57; ny=57; nz=67
vmin = 0; vmax = 600

# the format of input data is like following, 
# nz values separated by whitespace per line and nx*ny lines in total. 
# 0 1 2 ... nz-2 nz-1
# 0 1 2 ... nz-2 nz-1
datafile='temperature.dat'

# this list will contain all 3D data. 
xyz=[]

# each line contains nz elements of data. Read ny of nz elements, then increment in x-direction.
with open(datafile) as fin:
    iy = 0
    for data in fin:
        # initialize list yz every ny reads. 
        if (iy%ny==0):
            yz = []
        # data contains entire one line as a string. need to split with whitespace separator, 
        # then convert to float before appending yz. 
        d = data.split()
        d = [float(i) for i in d]
        yz.append(d)
        
        # every ny reads, list yz will have an x-sliced data, e.g. xyz[0][:][:], 
        # xyz[1][:][:], so on. append it to list xyz. 
        if (iy%ny==ny-1):
            xyz.append(yz)

        # increment the counter
        iy+=1

# set the upper/lower limists in the plot. 
norm = colors.Normalize(vmin=vmin, vmax=vmax)

for idx in range(nx):
    # create the figure
    fig = plt.figure(num=None, figsize=(10, 8), dpi=100, facecolor='w', edgecolor='k')

    # show the reference image
    ax1 = fig.add_subplot(111)

    print 'idx = ',idx
    
    # slice in x direction. set the upper/lower limits, 
    # then show color bar and the plot. 
    im1 = ax1.imshow(xyz[idx][:][:], interpolation='none', origin='lower')
    im1.set_norm(norm)
    plt.colorbar(im1)
    plt.show()
```
